# **Referencial Teórico**

A estruturação de um modelo de previsão de carga para o Sistema Interligado Nacional (SIN) exige a compreensão das dinâmicas que regem o consumo energético e das ferramentas estatísticas com capacidade de decifrá-las. Deste modo, este referencial teórico abordará estatísticas clássicas até as técnicas de aprendizagem de máquina.

**O Contexto da Carga de Energia no Brasil**

A complexidade da previsão no Brasil é acentuada por mudanças estruturais recentes. O Operador Nacional do Sistema Elétrico (ONS, 2026) aponta que a carga possui uma dependência crítica de fatores climáticos e econômicos, tornando a modelagem um alvo móvel. Além disso, a EPE (2024) destaca o crescimento da Micro e Minigeração Distribuída (MMGD) como um fator que altera os padrões tradicionais de demanda, exigindo modelos cada vez mais sensíveis a essas novas dinâmicas de consumo.

**Modelagem Estatística e Séries Temporais Clássicas**

A base da análise de séries temporais reside na capacidade de isolar componentes de tendência e sazonalidade. Nesse sentido, os modelos da família ARIMA são consolidados como padrão-ouro para dados que apresentam padrões sazonais repetitivos (OLIVEIRA; ABARRACIN; SILVA, 2024). O diferencial do SARIMA é a inclusão de parâmetros específicos $(P, D, Q, s)$ que tratam a sazonalidade anual, característica intrínseca ao consumo de energia no Brasil. De forma complementar, a aplicação prática em diferentes nichos, como o varejo, reforça que a eficácia desses modelos depende de um rigoroso processo de diferenciação e análise de funções de autocorrelação (ACF e PACF) para a definição correta dos parâmetros (OLIVEIRA, 2026b; 2026c). Além dos modelos autorregressivos, a Suavização Exponencial de Holt-Winters surge como uma alternativa robusta, permitindo capturar tanto tendências quanto sazonalidades de forma aditiva ou multiplicativa, dependendo da volatilidade da série observada (OLIVEIRA; ABARRACIN; SILVA, 2024).

**Evolução para Machine Learning e Variáveis Exógenas**

Embora os modelos clássicos sejam eficazes para séries univariadas, a carga do SIN é influenciada por fatores externos complexos. Pejuca (2026) argumenta que algoritmos de Machine Learning, como o XGBoost e Random Forest, superam as limitações de linearidade dos modelos ARIMA. Essas técnicas são particularmente superiores quando o modelo incorpora variáveis exógenas, como oscilações de temperatura, feriados ou indicadores do PIB.No entanto, a transição para essa abordagem exige uma engenharia de atributos (feature engineering) intensiva. De acordo com Pejuca (2026), a criação de lags (atrasos temporais) e médias móveis é essencial para que o modelo de aprendizado de máquina compreenda a dependência temporal que os modelos estatísticos já possuem de forma nativa.
